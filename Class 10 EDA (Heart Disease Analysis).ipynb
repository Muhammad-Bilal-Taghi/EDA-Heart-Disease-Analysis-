# f" (formating Script)

```python
plt.figure(figsize=(35,35))
for i, col in enumerate(categorical_value, 1):
    plt.subplot(4,4,i)
    sns.barplot(x=f"{col}",y='target',data=data)
    plt.ylabel("Possibility of having heart disease")
    plt.savefig("heartdisease.png")
    plt.xlabel(f"{col}")
```
